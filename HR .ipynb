### _Data Cleaning:_
---------------------------------
_Data cleaning is a critical preprocessing step in any data analysis project. It involves removing redundant columns, which helps in simplifying the dataset by eliminating columns that do not add value or are repetitive. Renaming columns ensures that column names are clear, consistent, and descriptive, facilitating easier understanding and manipulation. Dropping duplicates is necessary to eliminate duplicate rows, ensuring each record is unique and that the analysis is not skewed by repeated entries. Cleaning individual columns addresses inconsistencies within columns, corrects data types, and ensures that values are uniform. Handling missing values is also crucial, as it involves removing or imputing missing values to maintain the integrity of the dataset. These steps collectively ensure that the dataset is accurate, consistent, and ready for meaningful analysis._

# _Load the Dataset:_


```python
import pandas as pd

data = pd.read_csv('HR-Employee-Attrition.csv')
data.head()
```

# _Data Cleaning:_

```python
# Step 1: Deleting redundant columns
redundant_columns = ['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber']
data_cleaned = data.drop(columns=redundant_columns)
```

```python
# Check current column names
print(data_cleaned.columns)

# Example renaming (if needed, currently not necessary)
# data_cleaned.rename(columns={'OldName': 'NewName'}, inplace=True)
```

```python
# Step 3: Dropping duplicates
data_cleaned = data_cleaned.drop_duplicates()
```

```python
# Step 4: Cleaning individual columns (checking for inconsistencies and data types)
# Display basic info to identify columns with potential issues
print(data_cleaned.info())

# Handle specific column inconsistencies (if any)
# Example: data_cleaned['Column'] = data_cleaned['Column'].apply(some_cleaning_function)
```

```python
# Step 5: Removing NaN values
data_cleaned = data_cleaned.dropna()

# Verify the data after cleaning
data_cleaned.info()
data_cleaned.head()
```

# _Data Visualization:_

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### _Category-Based Analysis:_

```python
# Overtime count plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data_cleaned, x='OverTime')
plt.title('Count of Employees Working Overtime')
plt.show()
```

```python
# Marital Status count plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data_cleaned, x='MaritalStatus')
plt.title('Count of Employees by Marital Status')
plt.show()
```

```python
# Job Role count plot
plt.figure(figsize=(12, 8))
sns.countplot(data=data_cleaned, y='JobRole')
plt.title('Count of Employees by Job Role')
plt.show()
```

```python
# Gender count plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data_cleaned, x='Gender')
plt.title('Count of Employees by Gender')
plt.show()
```

```python
# Education Field count plot
plt.figure(figsize=(12, 8))
sns.countplot(data=data_cleaned, y='EducationField')
plt.title('Count of Employees by Education Field')
plt.show()
```

```python
# Department count plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data_cleaned, x='Department')
plt.title('Count of Employees by Department')
plt.show()
```

```python
# Business Travel count plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data_cleaned, x='BusinessTravel')
plt.title('Count of Employees by Business Travel Frequency')
plt.show()
```

### _Relations Between Variables:_

```python
# Overtime vs Age box plot
plt.figure(figsize=(12, 8))
sns.boxplot(data=data_cleaned, x='OverTime', y='Age')
plt.title('Overtime vs Age')
plt.show()
```

```python
# Total Working Years distribution plot
plt.figure(figsize=(10, 6))
sns.histplot(data=data_cleaned, x='TotalWorkingYears', bins=20, kde=True)
plt.title('Distribution of Total Working Years')
plt.show()
```

```python
# Education Level count plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data_cleaned, x='Education')
plt.title('Count of Employees by Education Level')
plt.show()
```

```python
# Number of Companies Worked distribution plot
plt.figure(figsize=(10, 6))
sns.histplot(data=data_cleaned, x='NumCompaniesWorked', bins=10, kde=True)
plt.title('Distribution of Number of Companies Worked')
plt.show()
```

```python
# Distance from Home distribution plot
plt.figure(figsize=(10, 6))
sns.histplot(data=data_cleaned, x='DistanceFromHome', bins=10, kde=True)
plt.title('Distribution of Distance from Home')
plt.show()
```
